```python
def Hello(name: str | None) -> str:
    if name is None:
        return "Hello, Guest!"
    return f"Hello, {name}!"

print(greet("apple"))
print(greet(None))
# print(Hello())
```

```python
def greet(user: str | None = None) -> str:
    return f"Hello, {user or 'Guest'}"
print(greet("apple"))
print(greet(None))
```

```python
def find_user(user_id: int) -> str | None:
    if user_id == 1:
        return "apple"
    return None

result1 = find_user(1)
result2 = find_user(99)
```

```python
users: dict[int, str] = {
    1: "apple",
    2: "banana"
}

def get_user(user_id: int) -> str | None:
    return users.get(user_id)
```

```python
print(get_user(1))
print(get_user(99))
```
