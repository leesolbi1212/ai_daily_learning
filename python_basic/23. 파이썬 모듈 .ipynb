# **1. 파일 종류 소개**
파이썬에서는 .ipynb 파일과 .py 파일 모두 파이썬 코드를 작성할 수 있지만, 사용 목적과 형식이 다릅니다.  

- .py :	Python 파일 :
일반적인 파이썬 코드 파일 (텍스트 기반)  
프로젝트 개발, 배포용 코드 작성
- .ipynb :	IPython Notebook 파일 :
Jupyter에서 사용하는 노트북 형식 파일 (JSON 기반)  
데이터 분석, 실험 기록, 강의용
