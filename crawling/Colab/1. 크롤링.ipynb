# **1. 크롤링**
크롤링(Crawling)이란 웹 크롤러(Web Crawler) 또는 스크레이퍼(Scraper)라고 불리는 프로그램이나 스크립트를 사용하여 인터넷상의 웹 페이지에서 데이터를 자동으로 수집하는 과정을 말합니다. 주로 검색 엔진이 웹사이트를 탐색하고 색인(Index)에 추가하기 위해 사용하는 기술이며, 특정 주제나 데이터를 수집해 분석하는 데에도 활용됩니다. 크롤링은 HTML, CSS, JavaScript로 구성된 웹페이지 구조를 파싱(Parsing)하여 원하는 정보를 추출하고, 이를 활용 가능한 데이터 형식으로 저장합니다.
> 크롤링 시에는 웹사이트의 이용 약관과 로봇 배제 표준(Robots.txt)을 준수하여 법적, 윤리적 문제를 방지해야 합니다.  
(사실 법적인 문제는 없지만, 윤리는 잘 지켜야죠?)


# **2. Basic English Speaking**
- [사이트] (https://basicenglishspeaking.com/daily-english-conversation-topics/)

```python
import requests #서버에 접속해서 html을 가져오는 역할
from bs4 import BeautifulSoup #가져온 데이터를 파싱한 후, 원하는 데이터를 빼오는 모듈
```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(site) #get방식으로 site에 접속
print(request) #객체자체를 일단 print 해봄 -> 정상적인 접속인지 아닌지 확인 가능
# <Response [200]> : 정상적인 접속
```

```python
request.text
```

```python
soup = BeautifulSoup(request.text, "html.parser") #".parser"에 뭘로 파싱할 지 정하면 됨
```

```python
box_div = soup.find('div',{'class':'thrv-columns'}) #find = 첫 번째로 걸리는 1개만 찾음, div이고, class가 thrv-columns인 것
print(box_div)
```

```python
links = box_div.find_all('a')
print(links) #리스트로 가져옴
```

```python
links[0].text
```

```python
for link in links:
    print(link.text)
```

```python
subject = []
for link in links:
    subject.append(link.text)
```

```python
print(len(subject))
print(subject)
```

```python
print(f'총 {len(subject)}개의 주제를 찾았습니다.')
for i in range (len(subject)):
    print(f'{i+1}. {subject[i]}')
```

# **3. 다음 뉴스 기사 제목 크롤링**

https://v.daum.net/v/20250715062440818

https://v.daum.net/v/20250715045805744

https://v.daum.net/v/20250715103243406

```python
def daum_news_title(news_id):
    url = f'https://v.daum.net/v/{news_id}'
    request = requests.get(url)
    soup = BeautifulSoup(request.text, "html.parser")
    title = soup.find('h3', {'class':'tit_view'})
    if title:
        return title.get_text().strip()
    return '제목없음'
```

```python
daum_news_title('20250715103243406')
```

```python
daum_news_title('20250715045805744')
```

```python
daum_news_title('20250715062440818')
```

# **4. 벅스 뮤직 차트**
- [사이트] ('https://music.bugs.co.kr/chart')

```python
request = requests.get('https://music.bugs.co.kr/chart')
print(request)
```

```python
soup = BeautifulSoup(request.text, "html.parser")
titles = soup.find_all('p',{'class':'title'})
# print(titles)
artists = soup.find_all('p',{'class':'artist'})
# print(artists)
```

```python
# 1. 노래제목 - 가수
for i, (t, a) in enumerate(zip(titles, artists), 1):
    title = t.text.strip()
    artist = a.text.strip().split('\n')[0]
    print(f'{i}. {title} - {artist}')
```

# **5. 멜론 차트**
- https://www.melon.com/chart/index.htm


```python
request = requests.get('https://www.melon.com/chart/index.htm')
print(request)
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} #웹사이트는 봇(bot)을 통한 자동 크롤링을 방지하기 위해 요청의 출처를 확인
request = requests.get('https://www.melon.com/chart/index.htm', headers=headers)
print(request)
```

```python
# 강사님 ver

soup = BeautifulSoup(request.text, "html.parser")
titles = soup.find_all('div',{'class':'rank01'})
# print(titles)
artists = soup.find_all('span',{'class':'checkEllipsis'})
# print(artists)
```

```python
for i, (t, a) in enumerate(zip(titles, artists), 1):
    title = t.text.strip()
    artist = a.text.strip()
    print(f'{i}. {title} - {artist}')
```

```python
titles = [tag.get_text(strip=True)
          for tag in soup.select('.ellipsis.rank01 a')]
artists = []
for div in soup.select('.ellipsis.rank02'):
    # 직접 div 자식인 <a> 태그만 선택 (span 안은 제외)
    names = [a.get_text(strip=True)
             for a in div.find_all('a', recursive=False)]
    # 중복 제거 (순서 유지)
    unique = list(dict.fromkeys(names))
    artists.append(", ".join(unique))
print(artists)
```

```python
for idx, (title, artist) in enumerate(zip(titles, artists), 1):
    print(f"{idx}. {title} - {artist}")
```

# **6. 네이버 증권 크롤링**

https://finance.naver.com/

```python
def naver_finance_info(code):
    url = f'https://finance.naver.com/item/main.naver?code={code}'
    request = requests.get(url)
    soup = BeautifulSoup(request.text, "html.parser")
    name_tag = soup.find('div', {'class':'wrap_company'}).find('a')
    name = name_tag.get_text(strip=True) if name_tag else '정보없음'
    price_tag = soup.find('p', {'class':'no_today'}).find('span', {'class':'blind'})
    price = price_tag.get_text(strip=True) if price_tag else '정보없음'
    vol_tag = soup.find('span', {'class':'sptxt sp_txt9'}).find('em')
    volume = vol_tag.get_text(strip=True) if vol_tag else '정보없음'

    return {
            "name":   name,
            "price":  price,
            "code":   code,
            "volume": volume
        }
```

```python
def naver_finance(code):
    site = f'https://finance.naver.com/item/main.naver?code={code}'
    request = requests.get(site)
    soup = BeautifulSoup(request.text, "html.parser")

    div_totalinfo = soup.find('div', {'class':'new_totalinfo'})
    name = div_totalinfo.find('h2').text.split('\n')[0]
    div_today = div_totalinfo.find('div', {'class':'today'})
    price = div_today.find('span', {'class':'blind'}).text
    price = price.replace(',', '')
    price = int(price)
    table_no_info = soup.find('table',{'class':'no_info'})
    tds = table_no_info.find_all('td')
    volume = tds[2].find('span',{'class':'blind'}).text
    volume = int(volume.replace(',', ''))
    dic = {"name":name, "price":price, "code":code, "volume":volume}
    return dic
```
